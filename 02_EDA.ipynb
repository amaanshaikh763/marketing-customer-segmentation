```python
import pandas as pd
import numpy as np

df = pd.read_csv("marketing_campaign.csv", sep="\t")
```

```python
df['Income'] = df['Income'].fillna(df['Income'].mean())

df['Age'] = 2025 - df['Year_Birth']

df['TotalSpend'] = (
    df['MntWines']
    + df['MntFruits']
    + df['MntMeatProducts']
    + df['MntFishProducts']
    + df['MntSweetProducts']
    + df['MntGoldProds']
)
```

```python
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

df['Age'].hist(bins=20)

plt.title('Customer Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Customers')

plt.show()
```

```python
df['Education'].value_counts().plot(
    kind='bar'
)

plt.title('Education Distribution')
plt.xlabel('Education')
plt.ylabel('Customer Count')

plt.show()
```

```python
plt.figure(figsize=(8,5))

df['Income'].hist(bins=20)

plt.title('Income Distribution')
plt.xlabel('Income')
plt.ylabel('Customers')

plt.show()
```

```python
df[['Income','TotalSpend']].head()
```

```python
df[['Income','TotalSpend']].corr()
```

```python
df.groupby('Education')['TotalSpend'].mean()
```

```python
df.groupby('Education')['TotalSpend'].mean().plot(
    kind='bar'
)

plt.title('Average Spending by Education')

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

plt.scatter(df['Income'], df['TotalSpend'])

plt.title('Income vs Total Spend')
plt.xlabel('Income')
plt.ylabel('Total Spend')

plt.show()
```
